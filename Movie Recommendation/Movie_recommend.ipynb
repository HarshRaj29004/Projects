```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
#MERGING MOVIES with CREDITS

movies = movies.merge(credits,on = "title")
```

```python
# ''' List of items or columns that will be considered :
#     1) generes
#     2) id
#     3) keywords
#     4) original_title
#     5) overview
#     6) cast
#     7) crew
# '''
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
#Droppping Rows Having Null values

movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
# Used to convert from string to list
import ast

#''' converting list of dict to list of genres '''
def convert(list1):
    list2 = []
    for i in ast.literal_eval(list1):
        list2.append(i["name"])
    return list2

movies['genres'] = movies['genres'].apply(convert)
    
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#Converting cast set
def convert2(list1):
    list2 = []
    t = 0
    for i in ast.literal_eval(list1):
        list2.append(i["character"])
        t+=1
        if(t == 3):
            break
    return list2
movies['cast'] = movies['cast'].apply(convert2)
```

```python
#Extracting director
def convert3(list1):
    l = []
    t = 0
    for i in ast.literal_eval(list1):
        if(i['job'] == 'Director'):
            l.append(i['name'])
            break
    return l
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
df = movies[['id','title','tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
df['tags'] = df['tags'].apply(lambda x: x.lower())
# df.head()
```

```python
# Stemming to vector --> making two words[make,making] same [make,make]
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stemming(str):
    l1 = []
    for i in str.split():
        l1.append(ps.stem(i))
    return " ".join(l1)

df['tags'] = df['tags'].apply(stemming)
```

```python
# Transforming dataframe to Vector
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')

vector = cv.fit_transform(df['tags']).toarray()

#print(cv.get_feature_names_out()) check the list words
```

```python
#Finding the Similarity between two movies
from sklearn.metrics.pairwise import cosine_similarity
similar = cosine_similarity(vector)
```

```python
#Recommendation 
# df[df['title'] == "Pirates of the Caribbean: At World's End"].index[0]
def recommend(df,name,similar):
    spot = df[df['title'] == name].index[0]
    similarity_scores = list(enumerate(similar[spot]))
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:6]
    # recommendations = []
    for i in sorted_scores:
        print(df.iloc[i[0]].title)

recommend(df,'Man of Steel',similar)
```
